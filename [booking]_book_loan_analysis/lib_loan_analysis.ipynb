# 16년, 17년 월별 도서 대출량(총합계)

- 아래 그래프를 살펴보면, **'독서의 계절'**이라 알려진 **가을(9, 10, 11월)**에 오히려 **가장 적은 대출량**을 기록한 것을 확인할 수 있다.
- 사람들이 도서관에서 책을 가장 많이 빌리는 계절은 **겨울**(그 중에서도 1월), 그리고 **여름**이다.(날씨, 그리고 방학의 영향?)
- 사람들은 날씨가 좋은 봄과 가을엔, 책을 읽기보다 밖에 나가 활동하는 것을 더 좋아하나보다.
- 신년 버프(1월)와 새학기 버프(3월)도 그래프에서 읽어낼 수 있다.

```python
plt.plot(pd.DataFrame(loan_16['총합계']))
plt.rcParams["figure.figsize"] = (12, 4)
plt.title(f'16년 월별 대출량(총합계)')
plt.ylabel('대출 건수')
plt.show()
```

```python
plt.plot(pd.DataFrame(loan_17['총합계']))
plt.rcParams["figure.figsize"] = (12, 4)
plt.title(f'17년 월별 대출량(총합계)')
plt.ylabel('대출 건수')
plt.show()
```

```python
loan_total_1617 = pd.concat([loan_16['총합계'], loan_17['총합계']])
```

```python
plt.plot(loan_total_1617)
plt.rcParams["figure.figsize"] = (20, 4)
plt.title(f'16-17년 월별 대출량(총합계)')
plt.ylabel('대출 건수')
plt.show()
```

- 16년과 비교해, 17년 도서 대출이 꽤나 많이 증가한 것을 알 수 있다.(하지만 월별 패턴은 유사하다.)

# 16년, 17년 월별 도서 대출량(연령대별)

```python
lgd = []
for k in list(loan_16.columns)[2:-1:3]:
    plt.plot(pd.DataFrame(loan_16[k]))
    lgd.append(k)
plt.rcParams["figure.figsize"] = (12, 8)
plt.title(f'16년 월별 대출량(연령대 별)')
plt.ylabel('대출 건수')
plt.legend(lgd)
plt.show()
```

```python
lgd = []
for k in list(loan_17.columns)[2:-1:3]:
    plt.plot(pd.DataFrame(loan_16[k]))
    lgd.append(k)
plt.rcParams["figure.figsize"] = (12, 8)
plt.title(f'17년 월별 대출량(연령대 별)')
plt.ylabel('대출 건수')
plt.legend(lgd)
plt.show()
```
